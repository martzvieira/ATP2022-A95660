```python
#TPC4

def inserirSala(cinema,sala):
    if sala not in cinema:
        cinema.append(sala)

    return cinema

def listar(cinema):
    for sala in cinema: #in só dá pra listas
        print(sala[2])

def disponivel(cinema,filme,lugar):
    disp = False
    for sala in cinema:
        if filme == sala[2]:
            if lugar not in sala[1]:
                disp = True

    return disp

def vendebilhete(cinema, filme, lugar):
    for sala in cinema:
        if filme == sala[2]:
            if lugar not in sala[1]:
                sala[1].append(lugar)

    return cinema

def listardisponibilidade(cinema):
    for sala in cinema:
        print(str(sala[2]) + ": " + str(sala[0] - len(sala[1])))

cinema1=[]

def menu():

    global cinema1

    print("1. Listar filmes")
    print("2. Inserir sala")
    print("3. Lugar disponível")
    print("4. Vender bilhete")
    print("5. Listar disponibilidade")

    opcao = int(input("Escolha opção"))

    if opcao == 1:
        listar(cinema1)
    elif opcao == 2:
        nlug = int(input("Número de lugares"))
        nfilme = str(input("Nome do filme"))
        sala1 = (nlug,[],nfilme)
        cinema1 = inserirSala(cinema1,sala1)
    elif opcao == 3:
        nlug = int(input("Número do lugar"))
        nfilme = str(input("Nome do filme"))
        if disponivel(cinema1,nfilme,nlug):
            print("Está disponível")
        else:
            print("Não está disponível")
    elif opcao == 4:
        nlug = int(input("Número do lugar"))
        nfilme = str(input("Nome do filme"))
        if disponivel(cinema1,nfilme,nlug):
            vendebilhete(cinema1,nfilme,nlug)
        else:
            print("Não está disponível")
    elif opcao == 5:
        listardisponibilidade(cinema1)

while(True):
    menu()
```
