```python
# alunos = [[id_aluno,nome,curso,tpc1,tpc2,tpc3,tpc4, media],.....]
# id = string

import matplotlib.pyplot as plt

def ler(ficheiro):
    
    f=open(str(ficheiro), "r")
    listaAlunos = []
    
    for line in f:
        aluno = line.split(',')
        listaAlunos.append(aluno)

    f.close()
    
    return listaAlunos

def distCurso(alunos):
    
    dici = {}

    for aluno in alunos:
        curso=aluno[2]
        if curso in dici:
            dici[curso] += 1  #adiciona de vez mais 1 à key que já existe
        else:
            dici[curso] = 1
    return dici

def calcmedia(alunos):
    
    for aluno in alunos:
        a = (aluno[3]+aluno[4]+aluno[5]+aluno[6])/4
        aluno.append(a)
    
    return aluno

def escaloes(alunos):
    for aluno in alunos:
        if aluno[7] >= 1 and aluno[7] <=4:
            escalao = 'E'
        
        elif aluno[7] <= 8:
            escalao = 'D'

        elif aluno[7] <= 12:
            escalao = 'C'
        
        elif aluno[7] <= 16:
            escalao = 'B'

        elif aluno[7] <= 20:
            escalao = 'A'

        aluno.append(escalao)

def distEscalao(alunos):

    dici = {}

    for aluno in alunos:
        escalao = aluno[8]

        if escalao in dici:
            dici[escalao] += 1
        
        else:
            dici[escalao] = 1

    return dici

def plot(dist):
    plt.plot(dist.keys(), dist.values())
    plt.show()
```
